# **`Identification Null with New Data`**

```python
#Identification Missing Value
import pandas as pd 
datamiss_csv = pd.read_csv("/shopping_data_missingvalue.csv")
print(datamiss_csv.isnull().values.any())
```

```python
#Identification Mean and missing value on dataset
print("Mean Value on Shopping Data")
print(datamiss_csv.mean())
print("Missing value on Dataset :")
print(datamiss_csv.head(10))
#Insert mean value into empty data
datamiss_csv = datamiss_csv.fillna(datamiss_csv.mean())
#missing value handled
print("missing value handled with mean :")
print(datamiss_csv.head(10))
```

```python
#Handle Missing Value With Median
import pandas as pd
datamiss_csv = pd.read_csv("/shopping_data_missingvalue.csv")
#Identification Median and missing value
print("Median Value on Shopping Data")
print(datamiss_csv.median())
print("Missing Value on Dataset")
print(datamiss_csv.head(10))
#Insert median value into empty data
datamiss_csv = datamiss_csv.fillna(datamiss_csv.median())
#missing value handled
print("missing value handled with median :")
print(datamiss_csv.head(10))
```
